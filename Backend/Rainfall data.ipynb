```python
#6 months

def generate_dates(start_months):
    date_list = []
    for year, months in start_months.items():
        for month in months:
            start_date = datetime(year, month, 1)
            if month == 12:
                end_date = datetime(year + 1, 1, 1) - timedelta(days=1)
            else:
                end_date = datetime(year, month + 1, 1) - timedelta(days=1)

            current_date = start_date
            while current_date <= end_date:
                date_list.append(current_date.strftime('%Y-%m-%d'))
                current_date += timedelta(days=1)
    return date_list

target_months = {
    2023: [11,12], 
    2024: [1,2,3,12], 
    2025: [1,2,3] 
}

breakdown_list = generate_dates(target_months)


all_data = []

# Iterate through each date
for current_date in breakdown_list:
    formatted_date = datetime.strptime(current_date, "%Y-%m-%d").strftime('%Y-%m-%d')
    api_url = f"https://api-open.data.gov.sg/v2/real-time/api/rainfall?date={formatted_date}"

    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Raise an error for bad status codes (4xx, 5xx)
        json_data = response.json()

        # Store response data
        all_data.append({ "date": formatted_date, "data": json_data['data']['stations'] })
        print(f" Data fetched for {formatted_date}")

    except requests.exceptions.RequestException as e:
        print(f" Error fetching data for {formatted_date}: {e}")

# Save all data to a JSON file
#with open("station_data_6months.json", "w", encoding="utf-8") as file:
#    json.dump(all_data, file, indent=4)

print("\n All data has been saved to 'rainfall_data_6mo.json' successfully!")

```
